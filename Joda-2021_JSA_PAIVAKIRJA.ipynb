<b>Datan kerääminen ja jalostaminen</b>

- Ryömijöiden ja raapijoiden läpikäynti.
- Tuotteistettu analytiikka tapahtuu tyypillisesti Business Intelligence (liiketoimintatiedon hallinta) organisaatiossa.
- ETL prosessi = extract, transform, load (Master data)
    - Systemaattinen aineisto, pitkät aikasarjat
    - Tuotteistettua tiedon hyödyntämistä organisaation toiminnassa
    - Saatavissa laajasti organisaatiossa
    - Helposti ladattavissa
- DAD prosessi = discover, access, distill
    - Samantyyppinen periaate kuin ETL, mutta:
        -Data-aineisto kerätään hyvin nopeasti
        -Analyysi tapahtuu nopeasti

<b>Data science workflow</b>

-	Ohjaava analytiikka jolloin datatiede on viety osaksi jonkin järjestelmän toiminnallisuutta

<b>Data-analytiikka liiketoiminnassa</b>

-	Yrityksistä vain noin kolmanneksella on data-analytiikan työkalut käytössä ja dataa hyödynnetään aktiivisesti

<b>Ryömintä</b>

-	Etsitään linkkejä joiden kautta voidaan hakea uusia linkkejä halutuilla parametreilla

<b>Raavinta</b>

-	Kerätään webbisivuilta kiinnostavaa tietoa, joka ollaan löydetty ryöminnän avulla
-	Raavinnassa on liiketoimintamielessä aina pitää mielessä tekijänoikeusasiat

<b>Tietotyypit</b>

-	Luvut (kokonaisluvut, floatit)
-	Teksti (stringit)
-	Boolet (True, False)
-	Objektit (sekalaista, voi olla myös sisäisiä listoja)
-	Aikatieto

<b>Data wrangling</b>

-	Tehdään datan siivoaminen käsin josta tekoäly voi tehdä automaattisen rutiinin jota on helppo käyttää jatkossa

<b>Data-aineiston tutkiminen</b>

-	Data-aineisto on hyödyllistä ensin siivota ja järjestää hyvin
-	Myös datan sisältöä ja rakennetta on hyvä tutkia ennen kuin varsinaisesti mitään aletaan kehittämään
